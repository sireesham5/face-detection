```python
import cv2
import os
face_path = os.path.join(cv2.data.haarcascades, "haarcascade_frontalface_default.xml")
eye_path = os.path.join(cv2.data.haarcascades, "haarcascade_eye.xml")
smile_path=os.path.join(cv2.data.haarcascades,'haarcascade_smile.xml')
smile_classifier=cv2.CascadeClassifier(smile_path)
face_classifier=cv2.CascadeClassifier(face_path)
eye_classifier=cv2.CascadeClassifier(eye_path)

def detect (gray,frame):
    faces=face_classifier.detectMultiScale(gray,scaleFactor=1.3, minNeighbors=5, minSize=(100, 100))
    for x,y,w,h in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),7)
        roi_gray=gray[y:y+h,x:x+w]
        roi_color=frame[y:y+h,x:x+w]
        eyes=eye_classifier.detectMultiScale(roi_gray,1.25,8,minSize=(40,30))
        smile=smile_classifier.detectMultiScale(roi_gray,1.2,8,minSize=(150,150))
        for ex,ey,ew,eh in eyes:
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),3)
        for sx,sy,sw,sh in smile:
            cv2.rectangle(roi_color,(sx,sy),(sx+sw,sy+sh),(0,0,255),3)
    return frame
video=cv2.VideoCapture(0)
while True:
    _,frame=video.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    gray = cv2.equalizeHist(gray)
    canvas=detect(gray,frame)
    cv2.imshow('videoo',canvas)
    if cv2.waitKey(1)&0xFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
